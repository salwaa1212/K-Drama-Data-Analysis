# Visualization & Exploratory Data Analysis

# Popular Genre & Themes

```python
# Splitting genres and themes/tags into separate columns
data[['Genres', 'Tags']] = data['Genre'].str.split(',', n=1, expand=True)

# Count the occurrences of each genre
genre_counts = data['Genres'].value_counts()

# Count the occurrences of each theme/tag
tag_counts = data['Tags'].value_counts()

# Plotting the top genres
plt.figure(figsize=(10, 6))
genre_counts.head(10).plot(kind='bar', color='blue')
plt.title('Top 10 Popular Genres')
plt.xlabel('Genres')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Plotting the top themes/tags
plt.figure(figsize=(10, 6))
tag_counts.head(10).plot(kind='bar', color='yellow')
plt.title('Top 10 Popular Themes/Tags')
plt.xlabel('Themes/Tags')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Checking the relationship between genres/themes and ratings
genre_ratings = data.groupby('Genres')['Rating'].mean()
tag_ratings = data.groupby('Tags')['Rating'].mean()

# Print the average rating for each genre
print("Average Ratings by Genre:")
print(genre_ratings)

# Print the average rating for each theme/tag
print("\nAverage Ratings by Theme/Tag:")
print(tag_ratings)
```

Popular Genre:
Based on Data Visualization, I have found that Korean Drama (K-Drama) with Genre = 'Thriller' is the Genre with the highest rating. Then, for Genre = 'Life', 'Horror', 'Romance' have the same rating value.

Popular Themes:
Based on Data Visualization, I have found that Korean Drama (K-Drama) with Themes/Tag = 'Comedy, Romance, Drama' is the Themes/Tag with the highest rating.

# Most Production per Year

```python
# Count the number of dramas produced each year
dramas_per_year = data['Year of release'].value_counts().sort_index()

# Plot the number of dramas produced each year horizontally
plt.figure(figsize=(10, 6))
dramas_per_year.plot(kind='barh', color='purple')
plt.title('Number of Dramas Produced Each Year')
plt.xlabel('Number of Dramas')
plt.ylabel('Year')
plt.show()
```

Based on Data Visualization, I have found that Korean Dramas produced from 2003 to 2021 have experienced ups and downs. However, this data for the years 2004 to 2011 is not available.

- From 2012 to 2013 there was an increase.
- From 2013 to 2015 there was a significant decrease.
- From 2015 to 2016 there was a drastic increase.
- From 2017 to 2018 there was a decrease.
- From 2018 to 2019 there was a drastic increase.
- From 2019 to 2021 there was a significant increase.

```python
# Count the number of dramas produced each year
dramas_per_year = data['Year of release'].value_counts().sort_index()

# Sort the data by number of productions in descending order
dramas_per_year_sorted = dramas_per_year.sort_values(ascending=True)

# Select the year with the highest number of productions
most_produced_year = dramas_per_year_sorted.index[0]

# Plot the number of dramas produced each year horizontally
plt.figure(figsize=(10, 6))
dramas_per_year_sorted.plot(kind='barh', color='purple')
plt.title('Number of Dramas Produced Each Year')
plt.xlabel('Number of Dramas')
plt.ylabel('Year')
plt.show()
```

Based on Data Visualization, I have found that the most Korean Dramas produced were in 2021 and the least in 2003.

# Drama with Highest/Lowest Rating in Action Genre

```python
# Filter data based on 'Action' genre
action_dramas = data[data['Genre'].str.contains('Action', case=False)]

# If 'Rating' is considered to be numeric, you can use the 'idxmax()' method to find the index of the row with the highest rating
highest_rated = action_dramas.loc[action_dramas['Rating'].idxmax()]

# Show the name of the drama with the highest rating
print("Name of Drama with Highest Rating in Action Genre:")
print('Names :', highest_rated['Name'])
print('Rating:', highest_rated['Rating'])

# If 'Rating' is considered to be numeric, you can use the 'idxmin()' method to find the index of the row with the lowest rating
lowest_rated = action_dramas.loc[action_dramas['Rating'].idxmin()]

# Show the name of the drama with the lowest rating
print("Name of Drama with Lowest Rating in Action Genre:")
print('Names :', lowest_rated['Name'])
print('Rating:', lowest_rated['Rating'])
```

Based on Data Visualization, I found that the Korean Drama (K-Drama) with Name = 'Happiness' is the K-Drama with the highest rating from Genre = 'Action'. Meanwhile, the Korean drama with the name = 'Rebel' is the K-Drama with the lowest rating from the Genre = 'Action'.

# Recommendation System

```python
# Sort dramas by rating
top_rated_dramas = data.sort_values(by='Rating', ascending=False)

# Get the top 10 highest rated dramas
top_10_dramas = top_rated_dramas.head(10)

# Print the top 10 highest rated dramas
print("Top 10 Highest Rated Dramas:")
print(top_10_dramas[['Name', 'Rating']])

# You can also suggest similar dramas based on genres or themes
# For example, recommend dramas with similar genres
similar_genre_dramas = data[data['Genres'].isin(top_10_dramas['Genres'].unique())]

# Print the top 5 similar dramas based on genres
print("\nTop 5 Similar Dramas based on Genres:")
print(similar_genre_dramas[['Name', 'Genres', 'Rating']].head(5))
```

Based on Data Visualization, I found that the Korean Drama with Name = 'Move to Heaven' and Genre = 'Life' is the K-Drama with the highest rating and is a recommendation to watch for Korean drama lovers who may have never watched the drama.
